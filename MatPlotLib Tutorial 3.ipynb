### Customizing Matplotlib with style sheets and rcParams

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.style.use('ggplot')
data = np.random.randn(50)
```

```python
# To list all available styles, use:
print(plt.style.available)
```

#### Defining your own style
You can create custom styles and use them by calling style.use, with the path or URL 
mpl_configdir/stylelib/presentation.mplstyle with the following:

```python
axes.titlesize : 24
axes.labelsize : 20
lines.linewidth : 3
lines.markersize : 10
xtick.labelsize : 16
ytick.labelsize : 16
```

```python
# To use said style configs
import matplotlib.pyplot as plt
plt.style.use('classic')
```

```python
# you can also use multiple style sheets that have different purposes:
import matplotlib.pyplot as plt
plt.style.use(['dark_background', 'classic'])
```

```python
# If you only want to use a style for a specific block of code, but 
# don't want to change the global styling, you can use the following:
with plt.style.context(('dark_background')):
    plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')
plt.show()
```

```python
# You can also change rc settings in a python script, or interactively
# from the python shell
mpl.rcParams['lines.linewidth'] = 2
mpl.rcParams['lines.color'] = 'r'
plt.plot(data)
```

```python
# Can also group these into the same string
mpl.rc('lines', linewidth=4, color='g')
plt.plot(data)
```
